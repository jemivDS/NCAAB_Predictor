```python
# https://www.basketball-reference.com/playoffs/NBA_2016.html
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022]

source = urllib.request.urlopen('https://www.basketball-reference.com/teams/{}/{}_games.html'.format('GSW',years[0])).read()
soup = BeautifulSoup(source,'lxml')

table = soup.find_all('table')
df = pd.read_html(str(table))[1]        

df
```

```python
histo = pd.DataFrame()


for tm in list(teamz):
    for szn in years:
        source = urllib.request.urlopen('https://www.basketball-reference.com/teams/{}/{}_games.html'.format(tm, szn)).read()
        soup = BeautifulSoup(source,'lxml')
        table = soup.find_all('table')
        df = pd.read_html(str(table))[0]
        
        df['teamgame'] = tm
        df['season'] = szn
        
        df = df[df['Opponent'] != 'Opponent'].reset_index(drop=True)
        df.drop(columns=['Unnamed: 3', "Notes"],inplace=True)

        histo = pd.concat([histo, df])
```

```python
histo.to_csv('NBA_2016_thru_2022_seasons.csv', index=False)
```
